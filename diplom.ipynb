```python
import numpy as np
from matplotlib.pyplot import plot
```

```python
def tf(p):
    p = 0 + p * 1j;
    m1 = 0.9;
    m2 = 1.1;
    beta = 0.03;
    f = - p ** 2 / (((p + beta) ** 2 + m1 ** 2) * ((p + beta) ** 2 + m2 ** 2));
    return(f)
```

```python
def lprsser200(w):
    reloc = 0;
    imloc = 0;
    iodd = -1;
    for k in range (1, 200): 
        iodd = -iodd;
        omk = k * w;
        reimloc = tf(omk);
        reloc = reloc + iodd * reimloc.real;
        if (iodd == 1):
            imloc = imloc + reimloc.imag / k;
    J = reloc + imloc * 1j;
    return(J)
```

```python
%matplotlib inline

xx = np.linspace(0.1, 2, 10000);
yy = np.zeros([10000]);
for i in range(1, 10000):
     yy[i] = lprsser200(xx[i]);
plot(yy)
```
